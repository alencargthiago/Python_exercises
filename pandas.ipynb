#### manipulação de dados ####

Também podemos adicionar e excluir novas clunas, modificar valores e calcular estatísticas, nesse caso iremos pegar o df para o ano de 2023 e adicionar uma nova coluna com uma condição


```python
import numpy as np
```

```python
#adicionando colunas
df['Ano_Analise'] =2023

df['Cobrança_Abusiva']= np.where(
    df['charges'] > 20000,
    "Sim",
    "Não"
)

df.head()
```

A função Where do numpy funciona como o if e elfse, ela irá verificar se uma condição é verdade, caso for verdadeiro, ela retornará 'Sim', caso contrário, ela retornará 'Não'. Nesse caso em específico, ela está verificando se a coluna charges é > 20000

 <b> Excluir colunas </b>

 Podemos utilizar o <b>df.drop ou inplace </b>. Qual a diferença? No df.drop ele irá apenas ocultar a visualização, mas não irá alterar o objeto no dataframe original. Por isso, é importante utilizar o inplace quando de fato você queira excluir definitivamente alguma informação.

```python
df.drop(columns="Ano_Analise")
```

tipo <br>
df.drop(columns="Ano_Analise", inplace=True)


Para atribuir novos valores, podemos utilizar a função map, nesse caso iremos substituir a coluna do dataframe que criamos momentos atrás, chamada cobrança abusiva e iremos susbtituir o Sim e o Não, por 1 e 0, da seguinte forma:


```python
#trocando sim por 1 e não por 0
df['Cobrança_Abusiva'] = df['Cobrança_Abusiva'].map({
    "Sim":1,
    "Não":0
})

#criando coluna com valor 2023

df["Ano_Analise"] = 2023

#Atualiza valor de 2023 para 2022

df["Ano_Analise"] = 2022

#exibindo

df.head()
```

```python
print(df['Cobrança_Abusiva'].unique())
```

```python
df['Cobrança_Abusiva'].fillna("Não Informado", inplace=True)
df['Cobrança_Abusiva'] = df['Cobrança_Abusiva'].map({
    "Sim": 1,
    "Não": 0,
    "Não Informado": 0  # ou qualquer valor padrão que faça sentido no seu contexto
})
#criando coluna com valor 2023

df["Ano_Analise"] = 2023

#Atualiza valor de 2023 para 2022

df["Ano_Analise"] = 2022

#exibindo

df.head()
```

<p><b>count:</b> Mostra o número de observações (linhas) não-nulas em cada coluna do DataFrame.</p>
<p><b>mean:</b> Calcula a média aritmética dos valores em cada coluna. É a soma de todos os valores dividida pelo número de valores.</p>
<p><b>std:</b> Representa o desvio padrão, que mede a quantidade de variação ou dispersão de um conjunto de valores. Um desvio padrão baixo indica que os valores tendem a estar próximos da média; um desvio padrão alto indica que os valores estão espalhados por uma gama mais ampla de valores.</p>
<p><b>min:</b> Mostra o valor mínimo em cada coluna do DataFrame.</p>
<p><b>25%:</b> Representa o primeiro quartil (Q1) dos dados, ou seja, 25% dos dados são menores que esse valor. É calculado ordenando os dados e tomando o valor no índice que é 25% do caminho através da lista.</p>
<p><b>50%:</b> Representa a mediana ou o segundo quartil (Q2) dos dados. Metade dos dados são menores que esse valor. É o valor no meio de um conjunto de dados ordenado.</p>
<p><b>75%:</b> Representa o terceiro quartil (Q3) dos dados, ou seja, 75% dos dados são menores que esse valor. É o valor abaixo do qual se encontram 75% dos dados.</p>
<p><b>max:</b> Mostra o valor máximo em cada coluna do DataFrame.</p>
