#### manipula√ß√£o de dados ####

Tamb√©m podemos adicionar e excluir novas clunas, modificar valores e calcular estat√≠sticas, nesse caso iremos pegar o df para o ano de 2023 e adicionar uma nova coluna com uma condi√ß√£o


```python
import numpy as np
```

```python
#adicionando colunas
df['Ano_Analise'] =2023

df['Cobran√ßa_Abusiva']= np.where(
    df['charges'] > 20000,
    "Sim",
    "N√£o"
)

df.head()
```

A fun√ß√£o Where do numpy funciona como o if e elfse, ela ir√° verificar se uma condi√ß√£o √© verdade, caso for verdadeiro, ela retornar√° 'Sim', caso contr√°rio, ela retornar√° 'N√£o'. Nesse caso em espec√≠fico, ela est√° verificando se a coluna charges √© > 20000

 <b> Excluir colunas </b>

 Podemos utilizar o <b>df.drop ou inplace </b>. Qual a diferen√ßa? No df.drop ele ir√° apenas ocultar a visualiza√ß√£o, mas n√£o ir√° alterar o objeto no dataframe original. Por isso, √© importante utilizar o inplace quando de fato voc√™ queira excluir definitivamente alguma informa√ß√£o.

```python
df.drop(columns="Ano_Analise")
```

Para atribuir novos valores, podemos utilizar a fun√ß√£o map, nesse caso iremos substituir a coluna do dataframe que criamos momentos atr√°s, chamada cobran√ßa abusiva e iremos susbtituir o Sim e o N√£o, por 1 e 0, da seguinte forma:


```python
#trocando sim por 1 e n√£o por 0
df['Cobran√ßa_Abusiva'] = df['Cobran√ßa_Abusiva'].map({
    "Sim":1,
    "N√£o":0
})

#criando coluna com valor 2023

df["Ano_Analise"] = 2023

#Atualiza valor de 2023 para 2022

df["Ano_Analise"] = 2022

#exibindo

df.head()
```

```python
df['Cobran√ßa_Abusiva'].fillna("N√£o Informado", inplace=True)
df['Cobran√ßa_Abusiva'] = df['Cobran√ßa_Abusiva'].map({
    "Sim": 1,
    "N√£o": 0,
    "N√£o Informado": 0  # ou qualquer valor padr√£o que fa√ßa sentido no seu contexto
})
#criando coluna com valor 2023

df["Ano_Analise"] = 2023

#Atualiza valor de 2023 para 2022

df["Ano_Analise"] = 2022

#exibindo

df.head()
```

<p><b>count:</b> Mostra o n√∫mero de observa√ß√µes (linhas) n√£o-nulas em cada coluna do DataFrame.</p>
<p><b>mean:</b> Calcula a m√©dia aritm√©tica dos valores em cada coluna. √â a soma de todos os valores dividida pelo n√∫mero de valores.</p>
<p><b>std:</b> Representa o desvio padr√£o, que mede a quantidade de varia√ß√£o ou dispers√£o de um conjunto de valores. Um desvio padr√£o baixo indica que os valores tendem a estar pr√≥ximos da m√©dia; um desvio padr√£o alto indica que os valores est√£o espalhados por uma gama mais ampla de valores.</p>
<p><b>min:</b> Mostra o valor m√≠nimo em cada coluna do DataFrame.</p>
<p><b>25%:</b> Representa o primeiro quartil (Q1) dos dados, ou seja, 25% dos dados s√£o menores que esse valor. √â calculado ordenando os dados e tomando o valor no √≠ndice que √© 25% do caminho atrav√©s da lista.</p>
<p><b>50%:</b> Representa a mediana ou o segundo quartil (Q2) dos dados. Metade dos dados s√£o menores que esse valor. √â o valor no meio de um conjunto de dados ordenado.</p>
<p><b>75%:</b> Representa o terceiro quartil (Q3) dos dados, ou seja, 75% dos dados s√£o menores que esse valor. √â o valor abaixo do qual se encontram 75% dos dados.</p>
<p><b>max:</b> Mostra o valor m√°ximo em cada coluna do DataFrame.</p>
